6. (10 points) A soccer team wins 60% of its games when it scores the first goal, and 10% of its games when the opposing team scores first. If the team scores the first goal about 30% of the time, what fraction of the games does it win?



100 games -> 60 times when it scores the first goal and 10 times when the opposing team scores

Team scores first about 30 times -> that implies the other 70 times opposite team scores first

P(W)=P(WF)P(F)+P(WF\`)*P(F\`)

P(W): Probability of winning a game.

P(F): Probability of scoring the first goal = 30/100

P(WS): Probability of winning given that the team scored the first goal = 60/100

P(F\`): Probability of opposite team scoring the first goal = 70/100

P(WF\`):  Probability of winning given that the opposite team scored the first goal = 10/100

So, based on formula ->

P(W) = (30/100 * 60/100) + (70/100 * 10/100) = 25/100 = 1/4
